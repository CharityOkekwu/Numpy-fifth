### 1. Stack plot:
#### Stack plots shows the whole data set with easy visualization of how each part contributes to the whole plot.
#### Example:


```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
months=np.array([2,3,4])
febuary=np.array([28])
march=np.array([31])
apirl=np.array([30])
plt.stackplot(months,febuary,march,apirl,labels=['Febuary','March','Apirl'])
plt.title('Stackplot')
plt.xlabel('Months')
plt.ylabel('Days of the month')
plt.legend()
plt.show
```

### AREA PLOT:
#### Area Plots are used to display quantitative data visually;
#### it also shows the rise and fall of various data series over time.

#### Example:

```python
import matplotlib.pyplot as plt
%matplotlib inline
week=[1,2,3,4,5]
shift1=[40,45,23,12,34]
shift2=[40,32,30,32,45]
shift3=[40,35,32,12,24]
plt.stackplot(week,shift1,shift2,shift3,labels=['Shift1','Shift2','Shift3'])
plt.xlabel('Weeks')
plt.legend()
```

### Sub Plots:
#### Sub plots are used to visualise multiple data sets all at once.
#### The sub plot function takes in 3 arguements, the row, column and index of plot.

```python
import matplotlib.pyplot as plt
x=([0,1,2,3,4])
y1=([50,20,30,10,40])
y2=([40,30,20,10,50])
plt.plot(x,y1,y2)
plt.show
```

#### Modification

```python
import matplotlib.pyplot as plt
x=([0,1,2,3,4])
y1=([50,20,30,10,40])
y2=([40,30,20,10,50])
#To represent each plot in a single figure, it'll be;
plt.subplot(1,2,1)
plt.plot(x,y1)

plt.subplot(1,2,2)
plt.plot(x,y2)

plt.show
```

#### Modification 2:


```python
import matplotlib.pyplot as plt
x=([0,1,2,3,4])
y1=([50,20,30,10,40])
y2=([40,30,20,10,50])
y3=([20,30,50,10,40])
y4=([10,20,30,40,50])
plt.suptitle("Subplot Examples")
#To represent each plot in a single figure and changing the dimension of the plot.
plt.subplot(2,2,1)
plt.plot(x,y1)
plt.title("First plot")


plt.subplot(2,2,2)
plt.plot(x,y2)
plt.title("Second plot")


plt.subplot(2,2,3)
plt.plot(x,y3)
plt.title("Third plot")


plt.subplot(2,2,4)
plt.plot(x,y4)
plt.title("Fourth plot")

plt.show
```

### Difference between correlation and causation.

#### Causation is a situation where one event affects the output/ocurrence of the other,
#### while correlation is simply a relationship between two variables, where action A relates to action B.

### What are outliers?

#### Outliers are values that fall more than three standard deviation from the mean.
#### Outliers can find their way into a dataset naturally through variability, human error,
#### faulty equipments or poor sampling.
#### Importance of outliers:
#### Outliers can hold useful information about data sets, they can also have effects on statistical results.
#### They can help you discover inconsistencies and detect any error in the statistical process.
